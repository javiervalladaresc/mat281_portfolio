EL conjunto de datos se denomina `Automobile_data.csv`, el cual contiene información tal como: compañia, precio, kilometraje, etc.

Lo primero es cargar el conjunto de datos y ver las primeras filas que lo componen:

```python
import pandas as pd
import os
import numpy as np
```

```python
# cargar datos
df = pd.read_csv(os.path.join("data","Automobile_data.csv")).set_index('index')
df.head()
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

1. Elimine los valores nulos (Nan)

```python
df = df[lambda df: df.notnull().all(axis=1)] #Realizamos un filtro para los valores nulos
```

2. Encuentra el nombre de la compañía de automóviles más cara

```python
comp = df.groupby("company") #Realizamos una agrupacion para las compañias
price = comp.agg({'price':[np.mean,sum]}).reset_index() #Calculamos el promedio y la suma de los precios de cada vehiculo 
precio = 0
indice = 0
for i in range(0,price.shape[0]-1): #Tomamos el precio más alto y vamos guardando el índice
    if price["price"]["mean"][i] >= precio:
        precio = price["price"]["mean"][i]
        indice = i
print("la compañia más cara es según el promedio de precio de venta es "+price["company"][indice])#Finalmente imprimimos el 
                                                                                                #auto más caro
```

3. Imprimir todos los detalles de Toyota Cars

```python
df_aux = df.loc[lambda x: x['company'] == "toyota"] #Filtramos los datos solo para la compañia toyota
df_aux.set_index("company")
```

4. Cuente el total de automóviles por compañía

```python
company = df["company"].value_counts() #Contamos las veces que aparece cada compañia y por ende cada auto 
company
```

5. Encuentra el coche con el precio más alto por compañía

```python
Car_max = dict() #Iniciamos un diccionario

for company in df["company"]: #Vamos estudiando compañia por compañia
     
    df_aux = df.loc[lambda x: x['company'] == company] #Filtramos los datos de cada compañia
    
    max_price = df_aux['price'].max() #Guardamos el precio más alto
    
    price_data = df_aux["price"] #Definimos una serie de precios
    
    car_data = df_aux["body-style"] #Definimos una serie con el tipo de arco
    
    A = price_data.index #Guardamos los indices de la serie precios
     
    for i in range(0,len(A)): #Iteramos y guardamos el auto más caro de cada compañia
        if price_data[A[i]] == max_price:
            car = car_data[A[i]],price_data[A[i]]
                 
    Car_max[company] = car #Asignamos el auto más caro para cada compañia
    
Cars = pd.DataFrame(Car_max)
Cars 
```

6. Encuentre el kilometraje promedio (**average-mileage**) de cada compañía automotriz

```python
comp = df.groupby("company") #Agrupamos por compañias
mileage = comp.agg({'average-mileage':[np.mean]}).reset_index() #Calculamos el promedio del kilometraje por cada compañia
mileage.set_index("company")
```

7. Ordenar todos los autos por columna de precio (**price**)

```python
df_price = df.sort_values("price") #Ordenamos la columna de precios de menor a mayor
df_price.set_index("price")
```

## Problema 02


Siguiendo la temática de los automóviles, resuelva los siguientes problemas:

#### a) Subproblema 01
A partir de los siguientes diccionarios:

```python
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'],
              'Price': [23845, 171995, 135925 , 71400]}
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 
                'Price': [29995, 23600, 61500 , 58900]}
```

* Cree dos dataframes (**carsDf1** y **carsDf2**) según corresponda.
* Concatene ambos dataframes (**carsDf**) y añada una llave ["Germany", "Japan"], según corresponda.

```python
carsDf1 = pd.DataFrame(GermanCars) #Definimos los dataframes
carsDf2 = pd.DataFrame(japaneseCars)

carsDf1["country"] = "Germany" #Agregamos una columna country correspondiente a cada diccionario
carsDf2["country"] = "Japan"
frames = [carsDf1,carsDf2]
result = pd.concat(frames) #Unimos ambos dataframe 
result
```

#### b) Subproblema 02

A partir de los siguientes diccionarios:

```python
Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}
```

* Cree dos dataframes (**carsDf1** y **carsDf2**) según corresponda.
*  Junte ambos dataframes (**carsDf**) por la llave **Company**.

```python
carsDf1 = pd.DataFrame(Car_Price) #Definimos ambos dataframes
carsDf2 = pd.DataFrame(car_Horsepower)

result = pd.merge(carsDf1, carsDf2, on='Company') #Pivoteamos ambos dataframe por la columna company
result
```
